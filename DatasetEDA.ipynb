##  Filling Missing Values

### 1) if missing values no in large number,we can directly drop them

```python
# Drop rows with missing values
df.dropna(axis=0).head()
```

```python
# Drop columns with missing values 
df.dropna(axis=1).head()
```

```python
df.dropna(how="all").head()  #drop only those rows who have all NaN values
```

```python
df.dropna(thresh=2).head()  #atleast 2 Nan
```

### 2) fill them with other values

```python
df.dtypes
```

```python
#we can fill with anything which we want like mean ,median etc
import numpy as np
mean=np.mean(df.priceRangeMin)
df.priceRangeMin.fillna(mean)
```

```python
df.fillna(method="ffill").head() #forward fill ..carry forward the previous value
```

```python
#we also have bfill ie backward fill...to fill value of next day
df.fillna(method="bfill").head()
```

####  #limit will control how many times a file must be copied

```python
df.fillna(method="ffill",limit=1)
```

```python
df.interpolate()  ##we can also give method 
```

### Third way is to replace them

```python
df.replace([-99999,-88888],np.nan)
```

```python
df.replace({
    -99999:np.nan,
    -88888:np.nan,
    'no available':'default'
})
```

### The graph is very difficult to analyse due to skewness,therefore converting values to log

```python
df['menus.amountMax'].fillna(np.mean(df['menus.amountMax']),inplace=True)
```

```python
#lets make groups
labels = ['0-99', '100-199', '200-299']
df['menus.amountMax-group'] =pd.cut(df['menus.amountMax'],range(0,301,100),right=False,labels=labels) 
```

# converting categorical data into numerical

```python
print(df.dtypes)
df['menus.name'] = df['menus.name'].astype('category')
print("\n after converting to category \n")
print(df.dtypes)
```

```python
#Then you can assign the encoded variable to a new column using the cat.codes accessor:
df["menus.name_cat"] = df['menus.name'].cat.codes
df.head()
```
